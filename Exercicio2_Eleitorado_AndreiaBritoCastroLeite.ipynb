## Item A - Classificação das variáveis

```python
import pandas as pd

#Classificação das Variaveis

resposta = [["cod_municipio_tse", "Qualitativa Nominal"],
            ["uf", "Qualitativa Nominal"],
            ["nome_municipio", "Qualitativa Nominal"],
            ["total_eleitores", "Quantitativa Discreta"],
            ["f_16", "Quantitativa Discreta"],
            ["f_17", "Quantitativa Discreta"],
            ["f_18_20", "Quantitativa Discreta"],
            ["f_21_24", "Quantitativa Discreta"],
            ["f_25_34", "Quantitativa Discreta"],
            ["f_35_44", "Quantitativa Discreta"],
            ["f_45_59", "Quantitativa Discreta"],
            ["f_60_69", "Quantitativa Discreta"],
            ["f_70_79", "Quantitativa Discreta"],
            ["f_sup_79", "Quantitativa Discreta"],
            ["gen_feminino", "Qualitativa Nominal"],
            ["gen_masculino", "Qualitativa Nominal"],
            ["gen_nao_informado", "Qualitativa Nominal"]]

resposta = pd.DataFrame(resposta, columns=["Variavel", "Classificação"])
resposta
```

image.png

### Item B - Tabela de frequência

Construa uma tabela de frequência para cada uma das variáveis qualitativas

```python
#Variaveis Qualitativas
qualitativas = resposta[resposta["Classificação"] == "Qualitativa Nominal"]

qualitativas.columns = ['Variável', 'Classificação', ]

# Exiba as variaveis quantitativas
print(qualitativas)
```

```python
##uf = Qualitativa Nominal

# Use a função value_counts() para calcular a tabela de frequência
# o metodo sort_values, para ordenar os valores, parêmetro "ascending" False, para vi em orderm decrescente
tabela_frequencia_uf = df['uf'].value_counts().sort_values(ascending=False).reset_index()

tabela_frequencia_uf.columns = ['Estado', 'Frequência', ]

# Tabela de frequência, exibindo quantos registros tem para casa estado.
print(tabela_frequencia_uf)
```

```python
#Soma dos eleitores por estado
#Comando agrupa todos os registros por estado e soma o total dos eleitores
soma_eleitores = df.groupby(['uf'])['total_eleitores'].sum().sort_values(ascending=False).reset_index()

soma_eleitores.columns = ['Estado', 'Soma_Eleitores']

# Exiba a tabela de frequência
print(soma_eleitores)
```

```python
#Municipio com mais números de eleitores
import pandas as pd

soma_eleitores_municipio = df.groupby(['nome_municipio'])['total_eleitores'].sum().sort_values(ascending=False)

soma_eleitores_municipio.columns = ['Municipio', 'Soma de Eleitores por municipio']
print(soma_eleitores_municipio)
```

```python
#Visualização de eleitores Feminino e Masculino por estado.
# Selecão das colunas
uf_genero = ['uf', 'gen_feminino', 'gen_masculino', 'gen_nao_informado']

# Crie um novo DataFrame com base nas colunas selecionadas
novo_df = df[uf_genero]

# Agrupe os dados pelo valor da coluna 'uf' e some as outras colunas
resultado_agrupado = novo_df.groupby('uf').sum().reset_index()

# Exiba o resultado
print(resultado_agrupado)
```

```python
#Percentual de eleitores por estado

dados_agrupados = df.groupby('uf')[['gen_feminino', 'gen_masculino', 'gen_nao_informado']].sum()

# Porcentagens para cada gênero em relação ao total por estado
total_por_estado = dados_agrupados.sum(axis=1)
dados_percentuais = (dados_agrupados / total_por_estado[:, None]) * 100

# Exiba os percentuais de eleitores por gênero para cada estado
print(dados_percentuais)
```

```python
# Estados em ordem decrescente com base no percentual de 'gen_feminino'
dados_percentuais = dados_percentuais.sort_values(by='gen_feminino', ascending=False)

# Exiba os percentuais de eleitores por gênero para cada estado em ordem decrescente
print(dados_percentuais)
```

```python
# Crie o gráfico de barras
plt.figure(figsize=(8, 6))  # Tamanho da figura (opcional)
plt.bar(soma_eleitores.Estado, soma_eleitores.Soma_Eleitores, color='royalblue')

# Ajuste automático do eixo y
plt.autoscale(enable=True, axis='y')

# Ajuste manual do eixo y
#plt.ylim(0, 50)  # Defina os valores mínimo e máximo do eixo y


# Adicione rótulos e título (opcional)
plt.xlabel('EStados')
plt.ylabel('Soma dos Eleitores')
plt.title('Gráfico de Barras - Estado com mais eleitores')

# Mostrar o gráfico de barras
plt.show()
```

## Estado de São Paulo

```python
import matplotlib.pyplot as plt
```

```python
# Filtrar o DataFrame para obter apenas os registros do estado 'SP'
df_sp = resultado_agrupado[resultado_agrupado['uf'] == 'SP']

# Somar os valores de 'gen_feminino', 'gen_masculino', 'gen_nao_informado'
total_feminino = df_sp['gen_feminino'].sum()
total_masculino = df_sp['gen_masculino'].sum()
total_nao_informado = df_sp['gen_nao_informado'].sum()

# Dados para o gráfico de pizza
valores = [total_feminino, total_masculino, total_nao_informado]
rotulos = ['Feminino', 'Masculino', 'Não informado']
cores = ['pink', 'lightblue', 'gray']

# Gráfico de pizza
plt.figure(figsize=(8, 8))  # Tamanho da figura (opcional)
plt.pie(valores, labels=rotulos, colors=cores, autopct='%1.1f%%', startangle=140)

# Título
plt.title('Distribuição de Gênero em SP')

# Mostrando o gráfico de pizza
plt.show()
```
